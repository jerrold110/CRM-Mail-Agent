```python
import pandas as pd
from sklearn.metrics import accuracy_score, precision_score, recall_score
```

```python


def get_df(func, columns, what_index):
    results_list = func()
    df = pd.DataFrame(results_list, columns=columns)
    df = df.set_index(what_index)

    print(df.dtypes)
    return df
```

# classify_email evaluation

```python
from evaluate import get_eval_results_email_classifier
classify_email_result_df = get_df(get_eval_results_email_classifier,
                                  ['test_no', 
                                   'output_topic',
                                   'output_urgency',
                                   'expected_topic',
                                   'expected_urgency'],
                                   'test_no')
```

```python
m1 = accuracy_score(classify_email_result_df['expected_topic'],
                    classify_email_result_df['output_topic'])
m2 = precision_score(classify_email_result_df['expected_urgency'],
                    classify_email_result_df['output_urgency'])
m3 = recall_score(classify_email_result_df['expected_urgency'],
                    classify_email_result_df['output_urgency'])

print(f'accuracy of email classifier topic is: {m1}')
print(f'precision of email classifier urgency is: {m2}')
print(f'recall of email classifier urgency is: {m3}')
print()
print(classify_email_result_df['output_topic'].value_counts())
print(classify_email_result_df['expected_urgency'].value_counts())
```

```python
# Accuracy
classify_email_result_df[classify_email_result_df['output_topic'] != classify_email_result_df['expected_topic']]
```

```python
# Recall
classify_email_result_df[(classify_email_result_df['expected_urgency'] == True) & (classify_email_result_df['output_urgency'] == False)]
```

```python
# Precision
classify_email_result_df[(classify_email_result_df['expected_urgency'] == False) & (classify_email_result_df['output_urgency'] == True)]
```

# get_comprehensive_product_query evaluation

```python
from evaluate import get_eval_results_get_comprehensive_product_query

get_product_query_df = get_df(get_eval_results_get_comprehensive_product_query,
                                  ['test_no', 
                                   'score',
                                   'explanation',
                                   'ground truth query',
                                   'generated query'],
                                   'test_no')
```

```python
import pprint

# Summary statistics for the 'score' column in get_product_query_df
s = get_product_query_df['score']

# Basic describe
describe = s.describe()

# Additional stats
median = s.median()
mode = s.mode().tolist()
missing = int(s.isna().sum())
value_counts = s.value_counts(dropna=False)

# Custom percentiles
percentiles = s.quantile([0.01, 0.05, 0.10, 0.25, 0.50, 0.75, 0.90, 0.95, 0.99]).to_dict()

# Perfect-score statistics (score == 1.0)
total = int(describe['count']) if 'count' in describe.index else len(s)
perfect_count = int((s == 1.0).sum())
perfect_percent = (perfect_count / total * 100) if total > 0 else float('nan')

# Aggregate results
summary = {
    'describe': describe.to_dict(),
    'median': median,
    'mode': mode,
    'percentiles': percentiles,
    'missing_values': missing,
    'value_counts': value_counts.to_dict(),
    'perfect_score_count': perfect_count,
    'perfect_score_percent': perfect_percent
}

# Print the summary
pprint.pprint(summary)
```
